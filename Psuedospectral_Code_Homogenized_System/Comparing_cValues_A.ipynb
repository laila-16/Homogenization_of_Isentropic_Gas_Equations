steps of finding the value of $$\langle  a^{-2}a_y [\![[\![ a]\!]]\!] \rangle$$  where   $~~~a(y)=cos(2\pi y)+cos(4\pi y)+sin(2\pi y) +5$
 1. find $\langle  a \rangle = \int_{0}^{1} a(y) dy$
 2. integrate $[\![ a]\!]= \int_{0}^{y} a(y) - \langle  a \rangle$
 3. integrate the result from step two again using the same limits of integration
 4. multiply the value in step 3 by $a^{-2}a_y$
 5. integrate 4 from $~0$ to $~~1$

```python
def double_bracket(f):
    # Compute the double bracket of a function
    mean = integrate.quad(f,0,1)[0]
    brace = lambda y: f(y)-mean
    brack_nzm = lambda y: integrate.quad(brace,0,y)[0]
    mean_bracket = integrate.quad(brack_nzm,0,1)[0]
    def brack(y):
        return integrate.quad(brace,0,y)[0] - mean_bracket
    return brack
```
